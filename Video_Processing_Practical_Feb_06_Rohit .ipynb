```python
import cv2

# Open a video file
cap = cv2.VideoCapture('birds flying.mp4')

while cap.isOpened():
    
    ret, frame = cap.read()
    if not ret:
        break
    # Process the frame (e.g., show it)
    cv2.imshow('Frame', frame)
    
    # Press 'q' to quit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
# Resized The Video
cap = cv2.VideoCapture('birds flying.mp4s')
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    # Resize frame
    resized_frame = cv2.resize(frame, (1000, 1000))
    cv2.imshow('Resized Frame', resized_frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```

```python
# Vidio Playing In Black Scale
cap = cv2.VideoCapture('birds flying.mp4')
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    # Convert frame to blackscale
    black_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    cv2.imshow('blac Frame', black_frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```

```python
# We Print The Text On The Video
cap = cv2.VideoCapture('birds flying.mp4')
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    # Add text
    cv2.putText(frame, 'This is birds flying Video creatdit by Rohit',
    (30, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255 ), 3, cv2.LINE_AA)
    cv2.imshow('Frame with Text', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
